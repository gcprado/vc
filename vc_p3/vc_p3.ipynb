Las formas localizadas tienen distintas características geométricas, que pueden estimarse a partir de sus contornos. Más información en la [documentación de OpenCV](https://docs.opencv.org/4.x/dd/d49/tutorial_py_contour_features.html).

```python
#Creación de polígono regular
def poligono_regular(image, ctr, radio, lados, color):
    pts = []
    ang_step = 2 * np.pi / lados
    for i in range(lados):
        ang = i * ang_step
        x = int(ctr[0] + radio * np.cos(ang))
        y = int(ctr[1] + radio * np.sin(ang))
        pts.append((x, y))
    pts = np.array(pts, np.int32)
    #regorganiza
    pts = pts.reshape((-1, 1, 2))
    cv2.fillPoly(image, [pts], color=color)

# Imagen vacía
img = np.zeros((400, 400, 1), dtype="uint8")
color = (255, 255, 255)

# Formas básicas
cv2.circle(img, (100, 100), 50, color, -1)  # Circular
poligono_regular(img, (250, 150), 50, 5, color)  # Polígono regular
cv2.ellipse(img, (200, 300), (100, 40), 0, 0, 360, color, -1)  # Elíptica

# Localiza contornos
contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Parámeros texto
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.3
thickness = 1

# Process each contour to calculate compactness and ellipse ratio (if possible)
for c in contours:
    # Puntos del contorno
    clon = len(c)

    # Área y perímetro
    area = cv2.contourArea(c)
    perimetro = cv2.arcLength(c, True)

    #Contenedor alineado con ejes de la imagen
    x,y,w,h = cv2.boundingRect(c)
    
    # Compactness: 4*pi*Area/Perimeter^2
    if perimetro > 0:
        compacidad = (perimetro ** 2) / area
    else:
        compactness = 0
    
    # Ajusta elipse si hay suficientes puntos
    if clon >= 5:
        elipse = cv2.fitEllipse(c)
        (center, axes, orientation) = elipse
        major_axis = max(axes)
        minor_axis = min(axes)
        elipse_ratio = major_axis / minor_axis
    else:
        elipse_ratio = None
    
    # Muestra valores en imageb
    cv2.putText(img, f"n: {clon:.1f}", (x, int(y+h+10)), font, font_scale, (255, 255, 255), thickness)
    cv2.putText(img, f"A: {area:.1f}", (x, int(y+h+20)), font, font_scale, (255, 255, 255), thickness)
    cv2.putText(img, f"P: {perimetro:.1f}", (x, int(y+h+30)), font, font_scale, color, thickness)
    cv2.putText(img, f"C: {compacidad:.1f}", (x, int(y+h+40)), font, font_scale, color, thickness)
    cv2.putText(img, f"ER: {elipse_ratio:.1f}", (x, int(y+h+50)), font, font_scale, color, thickness)
    
# Visualiza la imagen
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Formas")
plt.axis('off')
plt.show()
```

```python
#Carga imagen directamente en grises
imgorig = cv2.imread('./assets/microplasticos/MPs_test.jpg', cv2.IMREAD_GRAYSCALE)

img = cv2.GaussianBlur(imgorig,(5,5),0)

#Umbralizados
ret,imth1 = cv2.threshold(img,150,255,cv2.THRESH_BINARY_INV)
thotsu,imth2 = cv2.threshold(img,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
print('Umbral escogido ', thotsu)
imth3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
imth4 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
 
titles = ['Original', 'Fijo','Otsu th='+str(int(thotsu)),
            'Adaptivo promedio', 'Adaptivo Gaussiano']
images = [img, imth1, imth2, 255 - imth3, 255 - imth4]
 
for i in range(5):
    plt.subplot(2,5,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i], fontsize=7)
    plt.xticks([]),plt.yticks([])

    #Obtiene únicamente los contornos externos
    if i>0:
        res,imth = cv2.threshold(images[i],120,255,cv2.THRESH_BINARY)
        contornos, hierarchy= cv2.findContours(imth, 
        cv2.RETR_EXTERNAL , 
        cv2.CHAIN_APPROX_SIMPLE)  
        img_cont = np.zeros(img.shape)
        cv2.drawContours(img_cont, contornos, -1, (255,255,255), -1)  
        plt.subplot(2,5,i+6),plt.imshow(img_cont,'gray')
        plt.xticks([]),plt.yticks([])
plt.show()
```

¿Qué es una matriz de confusión?
Se utiliza para mostrar el comportamiento de un clasificador para las distintas clases conocidas, se relacionan las etiquetas de las muestras anotadas frente a las predichas por el clasificador. Se busca una matriz diagonal, pero la perfección es infrecuente.

El siguiente ejemplo, muestra el modo de obtener la matriz de confusión para un hipotético problema con cuatro clases, y valores de anotación (variable y) y predicción (variable y_pred) obtenidos de forma aleatoria.

Para ello se proporciona como imagen de test *MPs_test.jpg* y sus correpondientes anotaciones *MPs_test_bbs.csv* con la que deben obtener las métricas para su propuesta de clasificación de microplásticos, además de la matriz de confusión. La matriz de confusión permitirá mostrar para cada clase el número de muestras que se clasifican correctamente de dicha clase, y el número de muestras que se clasifican incorrectamente como perteneciente a una de las otras dos clases.

En el trabajo [SMACC: A System for Microplastics Automatic Counting and Classification](https://doi.org/10.1109/ACCESS.2020.2970498), las características geométricas utilizadas fueron:

- Área en píxeles
- Perímetro en píxeles
- Compacidad (relación entre el cuadrado del perímetro y el área de la partícula)
- Relación del área de la partícula con la del contenedor
- Relación del ancho y el alto del contenedor
- Relación entre los ejes de la elipse ajustada
- Definido el centroide, relación entre las distancias menor y mayor al contorno

```python
import cv2
import numpy as np
from pathlib import Path

# ==========================================================
# CONFIGURACIÓN INICIAL
# ==========================================================
imagenes_base = {
    "FRA": [
        "./assets/microplasticos/FRA.png",
        "./assets/microplasticos/fragment-03-olympus-10-01-2020.JPG"
    ],
    "PEL": [
        "./assets/microplasticos/PEL.png",
        "./assets/microplasticos/pellet-03-olympus-10-01-2020.JPG"
    ],
    "TAR": [
        "./assets/microplasticos/TAR.png",
        "./assets/microplasticos/tar-03-olympus-10-01-2020.JPG"
    ]
}

ruta_salida = Path("./assets/microplasticos/augmentation")
ruta_prueba = Path("./assets/microplasticos/MPs_test.jpg")
ruta_csv_prueba = Path("./assets/microplasticos/MPs_test_bbs.csv")
ruta_salida.mkdir(parents=True, exist_ok=True)

# ==========================================================
# FUNCIONES DE TRANSFORMACIÓN
# ==========================================================
def rotar_img(img, grados):
    """Rota una imagen en el ángulo indicado."""
    alto, ancho = img.shape[:2]
    centro = (ancho // 2, alto // 2)
    matriz = cv2.getRotationMatrix2D(centro, grados, 1.0)
    return cv2.warpAffine(img, matriz, (ancho, alto))


def modificar_brillo(img, intensidad):
    """Modifica el brillo multiplicando el canal de valor (V)."""
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV).astype(np.float32)
    hsv[..., 2] = np.clip(hsv[..., 2] * intensidad, 0, 255)
    return cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2BGR)


def agregar_ruido(img, sigma):
    """Aplica ruido gaussiano de nivel sigma."""
    ruido = np.random.normal(0, sigma, img.shape).astype(np.float32)
    img_ruidosa = np.clip(img.astype(np.float32) + ruido, 0, 255)
    return img_ruidosa.astype(np.uint8)


def desenfocar(img, k):
    """Desenfoca suavemente la imagen con un kernel cuadrado."""
    return cv2.GaussianBlur(img, (k, k), 0)


def generar_variantes(img, params):
    """
    Genera diferentes versiones aumentadas de una imagen original.
    Retorna una lista de (nombre, imagen_transformada)
    """
    versiones = [("base", img.copy())]

    # Espejados
    versiones += [
        ("flip_h", cv2.flip(img, 1)),
        ("flip_v", cv2.flip(img, 0))
    ]

    # Rotaciones
    for ang in params.get("rotaciones", []):
        versiones.append((f"rot_{ang}", rotar_img(img, ang)))

    # Brillo
    for i, f in enumerate(params.get("factores_brillo", [])):
        tipo = "mas_clara" if f > 1 else "mas_oscura"
        versiones.append((f"{tipo}_{i}", modificar_brillo(img, f)))

    # Ruido
    for i, n in enumerate(params.get("niveles_ruido", [])):
        versiones.append((f"ruido_{i}", agregar_ruido(img, n)))

    # Desenfoque
    for i, k in enumerate(params.get("tamaños_kernel", [])):
        versiones.append((f"blur_{i}", desenfocar(img, k)))

    return versiones
```

```python
# ==========================================================
# CREACIÓN DE VARIANTES DE IMÁGENES
# ==========================================================
parametros_aumentacion = {
    "rotaciones": [90, 180, 270],
    "factores_brillo": [0.8, 1.2],
    "niveles_ruido": [20],
    "tamaños_kernel": [5]
}

for etiqueta, lista_rutas in imagenes_base.items():
    for indice, ruta_img in enumerate(lista_rutas, start=1):
        imagen = cv2.imread(ruta_img)

        if imagen is None:
            print(f"⚠️ No fue posible abrir la imagen: {ruta_img}")
            continue

        variantes = generar_variantes(imagen, parametros_aumentacion)

        for j, (nombre_variante, img_resultante) in enumerate(variantes, start=1):
            nombre_archivo = f"{etiqueta}_{indice}_{nombre_variante}.png"
            destino = ruta_salida / nombre_archivo

            if destino.exists():
                print(f"⚠️ Sobrescribiendo archivo existente → {destino.name}")

            cv2.imwrite(str(destino), img_resultante)

print(f"\n✅ Proceso completado: imágenes aumentadas guardadas en '{ruta_salida}'")
```

```python
def obtener_caracteristicas(img):
    """
    Extrae características básicas de una imagen:
    - Brillo promedio (canal V en HSV)
    - Circularidad media ponderada por área
    - Contornos válidos (mayores a cierto tamaño)
    """
    if img is None or img.size == 0:
        return 0, 0, []

    # Conversión de color
    img_gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    brillo_promedio = np.mean(img_hsv[..., 2])

    # Umbralización adaptativa usando Otsu
    _, binaria = cv2.threshold(img_gris, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    binaria = cv2.morphologyEx(binaria, cv2.MORPH_CLOSE, np.ones((3, 3), np.uint8))

    contornos, _ = cv2.findContours(binaria, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    area_total = 0.0
    circularidad_acumulada = 0.0
    contornos_validos = []
    umbral_area = 100

    for cont in contornos:
        area = cv2.contourArea(cont)
        if area <= umbral_area:
            continue

        perimetro = cv2.arcLength(cont, True)
        if perimetro == 0:
            continue

        contornos_validos.append(cont)
        area_total += area

        circularidad = 4 * np.pi * area / (perimetro ** 2)
        circularidad_acumulada += circularidad * area

    circularidad_media = (circularidad_acumulada / area_total) if area_total > 0 else 0

    return brillo_promedio, circularidad_media, contornos_validos
```

```python
# ==========================================================
# VISUALIZACIÓN DE DETECCIÓN DE OBJETOS
# ==========================================================
fig, ejes = plt.subplots(1, len(imagenes_base), figsize=(18, 5))

for idx, (etiqueta, rutas) in enumerate(imagenes_base.items()):
    ruta_img = Path(rutas[0])

    if not ruta_img.exists():
        print(f"⚠️ Imagen no encontrada: {ruta_img}")
        continue

    imagen = cv2.imread(str(ruta_img))
    if imagen is None:
        print(f"⚠️ No se pudo leer la imagen: {ruta_img}")
        continue

    brillo, circ_media, contornos = obtener_caracteristicas(imagen)
    img_vista = imagen.copy()

    # Dibujar contornos detectados
    if contornos:
        cv2.drawContours(img_vista, contornos, -1, (0, 255, 255), 3)

    # Convertir BGR → RGB para visualizar correctamente
    img_rgb = cv2.cvtColor(img_vista, cv2.COLOR_BGR2RGB)
    ejes[idx].imshow(img_rgb)

    ejes[idx].set_title(f"{etiqueta} (Circularidad: {circ_media:.2f}), (Brillo: {brillo:.2f})", fontsize=14)
    ejes[idx].axis("off")

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

```python
import pandas as pd

# ==========================================================
# EXTRACCIÓN DE CARACTERÍSTICAS DE IMÁGENES AUMENTADAS
# ==========================================================
datos_caracteristicas = []

for etiqueta in imagenes_base.keys():
    archivos_img = sorted(ruta_salida.glob(f"{etiqueta}_*.png"))

    for ruta in archivos_img:
        img = cv2.imread(str(ruta))
        if img is None:
            continue

        brillo, circ_media, _ = obtener_caracteristicas(img)

        datos_caracteristicas.append({
            "Archivo": ruta.name,
            "Clase": etiqueta,
            "Brillo": brillo,
            "Circularidad": circ_media
        })

# Crear DataFrame
df_caracteristicas = pd.DataFrame(datos_caracteristicas)

# Cambiar nombres de columnas para mostrar
df_mostrar = df_caracteristicas.rename(columns={
    "Archivo": "Nombre de Archivo",
    "Clase": "Etiqueta",
    "Brillo": "Brillo (HSV)",
    "Circularidad": "Circularidad Ponderada"
})

# Aplicar estilo al DataFrame
styled_df = df_mostrar.style.set_table_styles([
    {"selector": "thead th", "props": [("background-color", "#4F81BD"), 
                                       ("color", "white"),
                                       ("font-weight", "bold"),
                                       ("text-align", "center")]},
    {"selector": "tbody td", "props": [("text-align", "center")]}
]).format({
    "Brillo (HSV)": "{:.3f}",
    "Circularidad Ponderada": "{:.3f}"
}).set_caption("🌟 Características Extraídas de Imágenes Aumentadas 🌟")

# Mostrar el DataFrame estilizado
display(styled_df)
```

```python
def clasificar_por_reglas(img, umbral_circ=0.78, umbral_brillo=182.5):
    """
    Clasifica una imagen según reglas simples:
    - Circularidad alta → pellet (PEL)
    - Brillo bajo → tar (TAR)
    - Caso contrario → fragmento (FRA)
    """
    brillo, circ_media, _ = obtener_caracteristicas(img)

    if circ_media > umbral_circ:
        return "PEL"
    elif brillo <= umbral_brillo:
        return "TAR"
    else:
        return "FRA"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix

# ==========================================================
# EVALUACIÓN DE PREDICCIÓN SOBRE IMAGEN DE PRUEBA
# ==========================================================
imagen_prueba = cv2.imread(ruta_prueba)
bbs_df = pd.read_csv(ruta_csv_prueba)

etiquetas_reales, etiquetas_predichas = [], []
datos_resultados = []

for idx, fila in bbs_df.iterrows():
    x1, y1, x2, y2 = map(int, [fila['x_min'], fila['y_min'], fila['x_max'], fila['y_max']])
    etiqueta_real = fila['label'].strip().upper()

    # Limitar coordenadas al tamaño de la imagen
    y1, y2 = max(0, y1), min(imagen_prueba.shape[0], y2)
    x1, x2 = max(0, x1), min(imagen_prueba.shape[1], x2)

    roi = imagen_prueba[y1:y2, x1:x2]
    if roi.size == 0 or roi.shape[0] <= 1 or roi.shape[1] <= 1:
        continue

    brillo, circ_media, _ = obtener_caracteristicas(roi)
    prediccion = clasificar_por_reglas(roi)

    etiquetas_reales.append(etiqueta_real)
    etiquetas_predichas.append(prediccion)

    datos_resultados.append({
        "Indice": idx,
        "Etiqueta_Real": etiqueta_real,
        "Etiqueta_Predicha": prediccion,
        "Brillo_HSV": brillo,
        "Circularidad_Ponderada": circ_media
    })

# Crear DataFrame con resultados
df_resultados = pd.DataFrame(datos_resultados)

# Renombrar columnas para mostrar
df_mostrar = df_resultados.rename(columns={
    "Indice": "Índice",
    "Etiqueta_Real": "Etiqueta Real",
    "Etiqueta_Predicha": "Etiqueta Predicha",
    "Brillo_HSV": "Brillo (HSV)",
    "Circularidad_Ponderada": "Circularidad Ponderada"
})

# Aplicar estilo al DataFrame
styled_df = df_mostrar.head(10).style.set_table_styles([
    {"selector": "thead th", "props": [("background-color", "#4F81BD"), 
                                       ("color", "white"),
                                       ("font-weight", "bold"),
                                       ("text-align", "center")]},
    {"selector": "tbody td", "props": [("text-align", "center")]}
]).format({
    "Brillo (HSV)": "{:.3f}",
    "Circularidad Ponderada": "{:.3f}"
}).set_caption("🌟 Resultados de Predicción sobre Imagen de Prueba 🌟")

# Mostrar accuracy
labels_objetivo = ["FRA", "PEL", "TAR"]
accuracy = accuracy_score(etiquetas_reales, etiquetas_predichas)
cm = confusion_matrix(etiquetas_reales, etiquetas_predichas, labels=labels_objetivo)

print(f"\n✅ Accuracy (Reglas simples): {accuracy*100:.2f}%")
display(styled_df)

# ==========================================================
# Visualización de predicciones sobre la imagen
# ==========================================================
colores = {"FRA": (0, 0, 255), "PEL": (0, 255, 0), "TAR": (255, 0, 0)}
img_vis = imagen_prueba.copy()

for i, fila in bbs_df.iterrows():
    if i >= len(etiquetas_predichas):
        continue
    x1, y1, x2, y2 = map(int, [fila['x_min'], fila['y_min'], fila['x_max'], fila['y_max']])
    x1, y1, x2, y2 = max(0, x1), max(0, y1), min(img_vis.shape[1], x2), min(img_vis.shape[0], y2)

    etiqueta_pred = etiquetas_predichas[i]
    color = colores.get(etiqueta_pred, (255, 255, 255))

    cv2.rectangle(img_vis, (x1, y1), (x2, y2), color, 2)
    cv2.putText(img_vis, etiqueta_pred, (x1, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(img_vis, cv2.COLOR_BGR2RGB))
plt.title("Predicciones sobre imagen de prueba")
plt.axis("off")
plt.show()

# ==========================================================
# Matriz de confusión
# ==========================================================
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=labels_objetivo, yticklabels=labels_objetivo)
plt.xlabel("Predicción")
plt.ylabel("Etiqueta real")
plt.title("Matriz de Confusión")
plt.show()
```
